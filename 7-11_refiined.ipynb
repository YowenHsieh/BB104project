```python
import requests
import json
import re
shops = []
area_dict = {
    "桃園市":['中壢區','平鎮區','龍潭區','楊梅區','新屋區','觀音區','桃園區','龜山區','八德區','大溪區','大園區','蘆竹區'],
    "新竹縣":["新埔鎮","關西鎮","竹北市","湖口鄉","新豐鄉","芎林鄉","竹東鎮","橫山鄉","北埔鄉","寶山鄉"],
    "新竹市":["東區","北區","香山區"],
    "苗栗縣":["竹南鎮","頭份市","南庄鄉","獅潭鄉","後龍鎮","通霄鎮","苑裡鎮","苗栗市","造橋鄉","頭屋鄉","公館鄉","大湖鄉","銅鑼鄉","卓蘭鎮","三灣鄉"]
}

for city in area_dict.keys():
    for district in area_dict[city]:
        
        data={
            "commandid":"SearchStore",
            "city":city,
            "town":district
            
                }
        shops += requests.post("http://emap.pcsc.com.tw/EMapSDK.aspx",data=data).text.split('<POIName>')

_id = re.findall('<POIID>(.{1,6})              </POIID>',str(shops))
name = re.findall(', \'(.{1,5})</POIName>',str(shops))
long = re.findall('<X>(.{1,20})</X>',str(shops))
lati = re.findall('<Y>(.{1,20})</Y>',str(shops))
tel = re.findall('<Telno>(.{1,20})        </Telno>',str(shops))
add = re.findall('<Address>(.{1,100})</Address>',str(shops))
wc = re.findall('<isLavatory>(.{1})</isLavatory>',str(shops))
atm = re.findall('<isATM>(.)</isATM>',str(shops))

storelist=[]

for i in range(1,len(name)):
    storelist.append({
        '_id':'uni_'+_id[i],
        'name':name[i],
        'address':add[i],
        'city':add[i][0:3],
        'district':add[i][3:6],
        'longitude':float(long[i])/1000000,
        'latitude':float(lati[i])/1000000,
        'tel':tel[i],
        'wc':wc[i],
        'atm':atm[i]
    })
stores=str(storelist).strip('[]')
stores

 
# for district in townshiptao:    
#     data= {
#         "commandid":"SearchStore",
#         "city":"桃園市",
#         "town":district
#     }
#     res = requests.post("http://emap.pcsc.com.tw/EMapSDK.aspx", data=data)
#     a=res.text.split('<POIName>')
#     storetao.append(a)
# storetao
```
