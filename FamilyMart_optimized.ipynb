```python
import requests
import json
from pymongo import MongoClient
final_shop_list = []
headers={
    "Referer":"http://www.family.com.tw/marketing/inquiry.aspx"
}

area_dict = {
    "桃園市":['中壢區','平鎮區','龍潭區','楊梅區','新屋區','觀音區','桃園區','龜山區','八德區','大溪區','大園區','蘆竹區'],
    "新竹縣":["新埔鎮","關西鎮","竹北市","湖口鄉","新豐鄉","芎林鄉","竹東鎮","橫山鄉"],
    "新竹市":["東區","北區","香山區"],
    "苗栗縣":["竹南鎮","頭份市","南庄鄉","獅潭鄉","後龍鎮","通霄鎮","苑裡鎮","苗栗市","造橋鄉","頭屋鄉","公館鄉","大湖鄉","銅鑼鄉","卓蘭鎮"]
}
shops_list =[]
for city in area_dict.keys():
    for district in area_dict[city]:
        shops = requests.get("http://api.map.com.tw/net/familyShop.aspx?searchType=ShopList&type=&city={}&area={}&road=&fun=showStoreList&key=6F30E8BF706D653965BDE302661D1241F8BE9EBC".format(city,district), headers=headers)
        shops_list.extend(json.loads(shops.text.split("showStoreList(")[1][:-1]))
for shop in shops_list:
    shop_dict = {
        '_id': "fm_" + (str(shop['SERID']).split(".")[0]),
        'city': shop["addr"][0:3],
        'district': shop["addr"][3:5] if shop["addr"][4] == "區" else shop["addr"][3:6],
        'name': shop['NAME'],
        'address': shop['addr'],
        'longitude': shop['px'],
        'latitude': shop['py'],
        'tel': shop['TEL'],
        'wc': "N"
    }
    if  "Toilet" in str(shop["all"]):
        shop_dict["wc"] = "Y"
    final_shop_list.append(shop_dict)
```

```python
client = MongoClient("localhost", 27017)
db = client["cuisines"]
db.cvs.insert_many(final_shop_list)
```
